```python
import pandas as pd
import warnings
from pycaret.classification import *
warnings.filterwarnings('ignore')
from sklearn import preprocessing
```

```python
df = pd.read_csv('Datasets/credit_score.csv')
```

```python
df=df[~(df.Total_Family_Members<df.Total_Children)]
```

```python
df.Family_Status.value_counts()
```

```python
x = df
y = df["Status"]
```

```python
cat_mask = x.dtypes==object
cat_cols = x.columns[cat_mask].tolist()
x.columns
```

```python
le = preprocessing.LabelEncoder()
x[cat_cols] = x[cat_cols].apply(lambda col: le.fit_transform(col))
x[cat_cols]
```

```python
x
```

```python
x.Job_Title.value_counts()
```

```python
x=x.drop(columns=["Status","Applicant_ID"],axis=1)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
X_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
X_resampled.shape
```

```python
df = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
df.Status.value_counts()
```

```python
df_for_modeling = df.sample(frac=0.7, random_state=123)
df_test = df.drop(df_for_modeling.index)

df_for_modeling.reset_index(inplace=True, drop=True)
df_test.reset_index(inplace=True, drop=True)

print('Data for Modeling: ' + str(df_for_modeling.shape))
print('Test Data For Predictions: ' + str(df_test.shape))
```

```python
stp = setup(data = df_for_modeling, target = 'Status',train_size = 0.8,
            fix_imbalance = True,
            session_id=123)
```

```python
best_model = compare_models()
```
